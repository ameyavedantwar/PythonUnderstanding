```python

def leap_year(year,flag):
    if(year % 4 != 0): #checking if it divisible by 4
        if(flag == 0):
            print("it is a common year")
        return 1       #this is a return to all the functions calling this 
                       #function stating that the passed year was a common or leap year
    if(year % 4 == 0 and year % 100 != 0):
        if(flag == 0):
            print("it is a leap year")
        return 0
    if(year % 100 == 0 and year % 400 != 0):
        if(flag == 0):
            print("it is a common year")
        return 1
    if(year % 400 == 0 and year % 3200 != 0):
        if(flag == 0):
            print("it is a leap year")
        return 0
    if(year % 400 != 0 and year % 3200 == 0):
        if(flag == 0):
            print("it is a common year")
        return 1
            

```

```python
def dayoftheweek(q,m,k,flag):
    if(m == "January" or m == "january"):
        m = 1 + 12
        j = k//100
        k = k%100
        k = k-1
    if(m == "February" or m == "february"):
        m = 2 + 12
        j = k//100
        k = k%100
        k = k-1
    if(m == "March" or m == "march"):
        m = 3
    if(m == "April" or m == "april"):
        m = 4
    if(m == "May" or m == "may"):
        m = 5
    if(m == "June" or m == "june"):
        m = 6
    if(m == "July" or m == "july"):
        m = 7
    if(m == "August" or m == "august"):
        m = 8
    if(m == "September" or m == "september"):
        m = 9
    if(m == "October" or m == "october"):
        m = 10
    if(m == "November" or m == "november"):
        m = 11
    if(m == "December" or m == "december"):
        m = 12
    j = k//100
    k = k%100
    m = (13*(m+1))//5
    k = k + (k//4) + (j//4)
    k = k - (2*j)
    h = (q + m + k) 
    h = h % 7
    d = ((h + 5) % 7) + 1
    h = int(h)
    d = int(d)
    if(flag == 0):
        if(h == 0 and d == 6 ):
            print("It is a Saturday")
        if(h == 1 and d == 7):
            print("It is a Sunday")
        if(h == 2 and d == 1):
            print("It is a Monday")
        if(h == 3 and d == 2):
            print("It is a Tuesday")
        if(h == 4 and d == 3):
            print("It is Wednesday")
        if(h == 5 and d == 4):
            print("It is Thursday")
        if(h == 6 and d == 5):
            print("It is Friday")
    return(d)
```
